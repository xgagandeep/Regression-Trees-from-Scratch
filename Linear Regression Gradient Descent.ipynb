```python
X = pd.read_csv("C:\\Users\\gagan\Desktop\\Hardwork pays off\\Training Data\\Linear_X_Train.csv")
y = pd.read_csv("C:\\Users\\gagan\Desktop\\Hardwork pays off\\Training Data\\Linear_Y_Train.csv")
Xtest = pd.read_csv("C:\\Users\\gagan\Desktop\\Hardwork pays off\\Test Cases\\Linear_X_Test.csv")

```

```python
X = X.values
y = y.values
```

```python
plt.scatter(X,y)
plt.show()
```

```python
#normalization
u = X.mean()
std = X.std()
X = (X - u)/std
plt.style.use("seaborn")
plt.scatter(X,y)
plt.show()
```

```python
def hypothesis(x,theta):
    y_ = theta[1]*x + theta[0]
    return y_
def error(X,Y,theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        x = X[i]
        y_ = hypothesis(x,theta)
        y = Y[i]
        total_error += (y_ - y)**2
        
    return total_error/m
```

```python
def gradient(X,Y,theta):
    grad = np.zeros((2,))
    m = X.shape[0]
    for i in range(m):
        x = X[i]
        y_ = hypothesis(x,theta)
        y = Y[i]
        grad[0] += (y_ - y)
        grad[1] += (y_ - y ) * x
    return grad/m





def gradientDescent(X,Y,max_iteration = 100,learning_rate = 0.1):
    theta = np.zeros((2,))
    error_list = []
    theta_list = []
    for i in range(max_iteration):
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        theta[0] = theta[0] - learning_rate * grad[0]
        theta[1] = theta[1] - learning_rate * grad[1]
        theta_list.append([theta[0],theta[1]])
    return error_list,theta,theta_list
```

```python
#Xpred = Xpred.values
error_list,theta,theta_list = gradientDescent(X,y)
Ypred = hypothesis(X,theta)
```

```python

def r2score():
    Ypred = hypothesis(X,theta)
    num = np.sum((y-Ypred)**2)
    denom = np.sum((y-y.mean())**2)
    score = (1-num/denom)
    return score*100
    
```
